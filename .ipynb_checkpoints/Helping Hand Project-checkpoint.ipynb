```python
from tkinter import *
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
import googletrans
from googletrans import Translator
import pyttsx3
import speech_recognition as sr
import customtkinter

customtkinter.set_appearance_mode("dark")
customtkinter.set_default_color_theme("green")

def speechtotext():
    try:
        recognizer = sr.Recognizer()

        def record_audio():
            with sr.Microphone() as source:
                print("Listening...")
                recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
                audio = recognizer.listen(source)
            return audio

        def recognize_speech():
            try:
                audio = record_audio()
                text = recognizer.recognize_google(audio, language=combo_language.get())
                print("Recognized text:", text)
                translated_text = translate_text(text)
                print("Translated text:", translated_text)
                text_output.delete(1.0, END)
                text_output.insert(END, translated_text)
                text_input.delete(1.0, END)
                text_input.insert(END, text)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
                messagebox.showerror("Error", "Could not connect to the internet.")

        def translate_text(text):
            translator = Translator()
            translated = translator.translate(text, dest=combo_translation_language.get())
            return translated.text

        root = Tk()
        root.title("Speech to Text")
        root.geometry("500x400")
        root.resizable(FALSE, FALSE)
        root.configure(background="White")

        language_var = StringVar()
        language_options = list(googletrans.LANGUAGES.values())
        combo_language = ttk.Combobox(root, textvariable=language_var, values=language_options)
        combo_language.pack(pady=10)
        combo_language.set("en")

        translation_language_var = StringVar()
        translation_language_options = list(googletrans.LANGUAGES.values())
        combo_translation_language = ttk.Combobox(root, textvariable=translation_language_var, values=translation_language_options)
        combo_translation_language.pack(pady=10)
        combo_translation_language.set("en")

        record_button = Button(root, text="Record", command=recognize_speech)
        record_button.pack(pady=5)

        Label(root, text="Recognized Text:").pack()
        text_input = Text(root, height=5, width=50)
        text_input.pack(pady=5)

        Label(root, text="Translated Text:").pack()
        text_output = Text(root, height=5, width=50)
        text_output.pack(pady=5)

    except Exception as e:
        print("Error:", e)
        messagebox.showerror("Error", "An error occurred. Please try again.")

app = customtkinter.CTk()
app.title("Helping Hand")
app.geometry("700x500")

frame = customtkinter.CTkFrame(master=app, corner_radius=20,)
Label(frame, text="Helping Hands", font="ariel 25 bold", fg="navy", bg="grey").pack(fill="both")
frame.pack(pady=20, padx=20, fill="both", expand=True)

button_1 = customtkinter.CTkButton(master=frame,
                                   text="Text-to-Text",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=texttrans,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_1.pack(padx=20, pady=20)

button_2 = customtkinter.CTkButton(master=frame,
                                   text="Text-to-Speech ",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   compound="right",
                                   command=textspeech,
                                   anchor="e",
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E",
                                  )

button_2.pack(padx=20, pady=20)

button_3 = customtkinter.CTkButton(master=frame,
                                   text="Speech-to-Text",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=speechtotext,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_3.pack(padx=20, pady=20)

button_4 = customtkinter.CTkButton(master=frame,
                                   text="File-to-Speech",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=filespeech,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_4.pack(padx=20, pady=20)

app.mainloop()
```

```python
from tkinter import *
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
import googletrans
from googletrans import Translator
import pyttsx3
import speech_recognition as sr
import customtkinter

customtkinter.set_appearance_mode("dark")
customtkinter.set_default_color_theme("green")

app = customtkinter.CTk()
app.title("Helping Hand")
app.geometry("800x600")

frame = customtkinter.CTkFrame(master=app, corner_radius=20)
Label(frame, text="Helping Hands- A Dynamic Transaltor", font="ariel 25 bold", fg="navy", bg="grey").pack(fill="both")
frame.pack(pady=20, padx=20, fill="both", expand=True)

def texttrans():
    root = Tk()
    root.title("Google Translator")
    root.geometry("1080x400")
    root.resizable(False, False)
    root.configure(background="White")

    def label_change():
        c = combo1.get()
        c1 = combo2.get()
        label1.configure(text=c.upper())
        label2.configure(text=c1.upper())
        root.after(1000, label_change)
        
    def translate_now():
        text_ = text1.get(1.0, END)
        t1 = Translator()
        trans_text = t1.translate(text_, dest=combo2.get(), src=combo1.get())
        trans_text = trans_text.text
        
        text2.delete(1.0, END)
        text2.insert(END, trans_text)

    language = googletrans.LANGUAGES
    languageV = list(language.values())
    lang1 = language.keys()

    # First combo box
    combo1 = ttk.Combobox(root, values=languageV, font="Roboto 14", state="r")
    combo1.place(x=110, y=20)
    combo1.set("ENGLISH")

    label1 = Label(root, text="ENGLISH", font="segoe 30 bold", bg="white", width=18, bd=5, relief=GROOVE)
    label1.place(x=10, y=50)

    # Second combo box
    combo2 = ttk.Combobox(root, values=languageV, font="Roboto 14", state="r")
    combo2.place(x=730, y=20)
    combo2.set("SELECT LANGUAGE")

    label2 = Label(root, text="ENGLISH", font="segoe 30 bold", bg="white", width=18, bd=5, relief=GROOVE)
    label2.place(x=620, y=50)

    # First frame
    f = Frame(root, bg="Black", bd=5)
    f.place(x=10, y=118, width=440, height=210)

    text1 = Text(f, font="Roboto 20", bg="white", relief=GROOVE, wrap=WORD)
    text1.place(x=0, y=0, width=430, height=200)

    scrollbar1 = Scrollbar(f)
    scrollbar1.pack(side="right", fill='y')

    scrollbar1.configure(command=text1.yview)
    text1.configure(yscrollcommand=scrollbar1.set)

    # Second frame
    f1 = Frame(root, bg="Black", bd=5)
    f1.place(x=620, y=118, width=440, height=210)

    text2 = Text(f1, font="Roboto 20", bg="white", relief=GROOVE, wrap=WORD)
    text2.place(x=0, y=0, width=430, height=200)

    scrollbar2 = Scrollbar(f1)
    scrollbar2.pack(side="right", fill='y')

    scrollbar2.configure(command=text2.yview)
    text2.configure(yscrollcommand=scrollbar2.set)

    # Translate button
    translate = Button(root, text="Translate", font=("Roboto", 15), activebackground="white", cursor="hand2", bd=1,
                       width=10, height=2, bg="black", fg="white", command=translate_now)
    translate.place(x=476, y=150)

    label_change()
    
def textspeech():
    try:
        def speaknow():
            engine = pyttsx3.init()
            engine.setProperty('rate', 150)
            engine.setProperty('voice', combo_voice.get())
            engine.say(text1.get("1.0", END))
            engine.runAndWait()
            engine.stop()
      
        root = Tk()
        root.title("Text to speech") 
        root.geometry("400x200") 
        root.resizable(False, False) 

        obj = LabelFrame(root, text="Text to speech", font=20, bd=1)
        obj.pack(fill="both", expand="yes", padx=10, pady=10)

        lbl = Label(obj, text="Text", font=30) 
        lbl.pack(side=tk.LEFT, padx=5)
        text1 = Text(obj, font=30, width=25, bd=5) 
        text1.pack(side=tk.LEFT, padx=10)

        voice_var = StringVar()
        voices = pyttsx3.init().getProperty('voices')
        voice_options = [voice.name for voice in voices]
        combo_voice = ttk.Combobox(obj, textvariable=voice_var, values=voice_options)
        combo_voice.pack(side=tk.LEFT, padx=10)
        combo_voice.set(voice_options[0])
        
        def speak():
            speaknow()
            
        btn_speak = Button(obj, text="Speak", font=20, bg="black", fg="white", command=speak) 
        btn_speak.pack(side=tk.LEFT, padx=10) 
        
        def on_enter(event):
            speaknow()
        
        text1.bind("<Return>", on_enter)
        
    except Exception as e:
        print("Error:", e)

def filespeech():
    try:
        root = Tk()
        root.title("File to Speech") 
        root.geometry("500x300")
        root.resizable(False, False)
        root.configure(background="White")

        lbl = Label(root, text="Select a text file", bg="white")
        lbl.place(x=10, y=50)

        def file():
            file_path = filedialog.askopenfilename()
            if file_path:
                with open(file_path, "r") as file:
                    text = file.read()
                engine = pyttsx3.init()
                engine.setProperty('rate', 150)
                engine.setProperty('voice', combo_voice.get())
                engine.say(text)
                engine.runAndWait()
                engine.stop()
        
        btn_browse = Button(root, text="Browse", command=file)
        btn_browse.place(x=150, y=50)

        voice_var = StringVar()
        voices = pyttsx3.init().getProperty('voices')
        voice_options = [voice.name for voice in voices]
        combo_voice = ttk.Combobox(root, textvariable=voice_var, values=voice_options)
        combo_voice.place(x=250, y=50)
        combo_voice.set(voice_options[0])
        
    except Exception as e:
        print("Error:", e)

def speechtotext():
    try:
        recognizer = sr.Recognizer()

        def record_audio():
            with sr.Microphone() as source:
                print("Listening...")
                recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
                audio = recognizer.listen(source)
            return audio

        def recognize_speech():
            try:
                audio = record_audio()
                src_language = combo_language.get()
                src_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(src_language.lower())]
                text = recognizer.recognize_google(audio, language=src_language_code)
                print("Recognized text:", text)
                text_output.delete(1.0, END)
                text_output.insert(END, text)
                
                # Translate recognized text if translation language selected
                if combo_translate_language.get() != "None":
                    dest_language = combo_translate_language.get()
                    dest_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(dest_language.lower())]
                    translator = Translator()
                    translated_text = translator.translate(text, dest=dest_language_code)
                    text_translated.delete(1.0, END)
                    text_translated.insert(END, translated_text.text)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
                messagebox.showerror("Error", "Could not connect to the internet.")

        root = Tk()
        root.title("Speech to Text")
        root.geometry("700x600")
        root.resizable(False, False)
        root.configure(background="White")

        # Label for recognized speech
        label_recognized = Label(root, text="Recognized speech", font=("Arial", 14, "bold"))
        label_recognized.pack(pady=10)

        # Language selection for speech recognition
        language_var = StringVar()
        language_options = list(googletrans.LANGUAGES.values())
        combo_language = ttk.Combobox(root, textvariable=language_var, values=language_options)
        combo_language.pack(pady=5)
        combo_language.set("english")

        # Label for translated speech
        label_translated = Label(root, text="Translated speech", font=("Arial", 14, "bold"))
        label_translated.pack(pady=10)

        # Language selection for translation
        translate_language_var = StringVar()
        translate_language_options = ["None"] + language_options
        combo_translate_language = ttk.Combobox(root, textvariable=translate_language_var, values=translate_language_options)
        combo_translate_language.pack(pady=5)
        combo_translate_language.set("None")

        record_button = Button(root, text="Record", command=recognize_speech)
        record_button.pack(pady=5)

        text_output = Text(root, height=10, width=50)
        text_output.pack(pady=10)
        
        text_translated = Text(root, height=10, width=50)
        text_translated.pack(pady=10)

    except Exception as e:
        print("Error:", e)
        messagebox.showerror("Error", "An error occurred. Please try again.")

def speechtospeech():
    try:
        recognizer = sr.Recognizer()
        engine = pyttsx3.init()

        def record_audio():
            with sr.Microphone() as source:
                print("Listening...")
                recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
                audio = recognizer.listen(source)
            return audio

        def speaknow(text):
            engine.say(text)
            engine.runAndWait()
            engine.stop()

        def recognize_and_translate_speech():
            try:
                audio = record_audio()
                src_language = combo_language.get()
                src_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(src_language.lower())]
                text = recognizer.recognize_google(audio, language=src_language_code)
                print("Recognized text:", text)
                text_output.delete(1.0, END)
                text_output.insert(END, text)
                
                # Speak recognized text
                speaknow(text)
                
                # Translate recognized text
                dest_language = combo_translate_language.get()
                dest_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(dest_language.lower())]
                translator = Translator()
                translated_text = translator.translate(text, dest=dest_language_code).text
                print("Translated text:", translated_text)
                text_translated.delete(1.0, END)
                text_translated.insert(END, translated_text)
                
                # Speak translated text
                speaknow(translated_text)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
                messagebox.showerror("Error", "Could not connect to the internet.")

        root = Tk()
        root.title("Speech to Speech")
        root.geometry("700x500")
        root.resizable(False, False)
        root.configure(background="White")

        # Label for recognized speech
        label_recognized = Label(root, text="Speech to Speech Translation", font=("Arial", 14, "bold"))
        label_recognized.pack(pady=10)

        # Language selection for speech recognition
        language_var = StringVar()
        language_options = list(googletrans.LANGUAGES.values())
        combo_language = ttk.Combobox(root, textvariable=language_var, values=language_options)
        combo_language.pack(pady=5)
        combo_language.set("english")

        # Language selection for translation
        translate_language_var = StringVar()
        translate_language_options = language_options
        combo_translate_language = ttk.Combobox(root, textvariable=translate_language_var, values=translate_language_options)
        combo_translate_language.pack(pady=5)
        combo_translate_language.set("spanish")

        record_button = Button(root, text="Translate Speech", command=recognize_and_translate_speech)
        record_button.pack(pady=20)

        text_output = Text(root, height=10, width=50)
        text_output.pack(pady=10)
        
        text_translated = Text(root, height=10, width=50)
        text_translated.pack(pady=10)

    except Exception as e:
        print("Error:", e)
        messagebox.showerror("Error", "An error occurred. Please try again.")

button_1 = customtkinter.CTkButton(master=frame,
                                   text="Text-to-Text",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=texttrans,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_1.pack(padx=20, pady=20)

button_2 = customtkinter.CTkButton(master=frame,
                                   text="Text-to-Speech",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=textspeech,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E",
                                  )

button_2.pack(padx=20, pady=20)

button_3 = customtkinter.CTkButton(master=frame,
                                   text="Speech-to-Text",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=speechtotext,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_3.pack(padx=20, pady=20)

button_4 = customtkinter.CTkButton(master=frame,
                                   text="File-to-Speech",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=filespeech,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_4.pack(padx=20, pady=20)

button_5 = customtkinter.CTkButton(master=frame,
                                   text="Speech-to-Speech",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=speechtospeech,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_5.pack(padx=20, pady=20)

app.mainloop()
```

```python
from tkinter import *
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
import googletrans
from googletrans import Translator
import pyttsx3
import speech_recognition as sr
import customtkinter

customtkinter.set_appearance_mode("dark")
customtkinter.set_default_color_theme("green")

app = customtkinter.CTk()
app.title("Helping Hand")
app.geometry("800x600")

frame = customtkinter.CTkFrame(master=app, corner_radius=20)
Label(frame, text="Helping Hands- A Dynamic Translator", font="ariel 25 bold", fg="navy", bg="grey").pack(fill="both")
frame.pack(pady=20, padx=20, fill="both", expand=True)

def texttrans():
    root = Tk()
    root.title("Google Translator")
    root.geometry("1080x400")
    root.resizable(False, False)
    root.configure(background="White")

    def label_change():
        c = combo1.get()
        c1 = combo2.get()
        label1.configure(text=c.upper())
        label2.configure(text=c1.upper())
        root.after(1000, label_change)
        
    def translate_now():
        text_ = text1.get(1.0, END)
        t1 = Translator()
        trans_text = t1.translate(text_, dest=combo2.get(), src=combo1.get())
        trans_text = trans_text.text
        
        text2.delete(1.0, END)
        text2.insert(END, trans_text)

    language = googletrans.LANGUAGES
    languageV = list(language.values())
    lang1 = language.keys()

    # First combo box
    combo1 = ttk.Combobox(root, values=languageV, font="Roboto 14", state="r")
    combo1.place(x=110, y=20)
    combo1.set("ENGLISH")

    label1 = Label(root, text="ENGLISH", font="segoe 30 bold", bg="white", width=18, bd=5, relief=GROOVE)
    label1.place(x=10, y=50)

    # Second combo box
    combo2 = ttk.Combobox(root, values=languageV, font="Roboto 14", state="r")
    combo2.place(x=730, y=20)
    combo2.set("SELECT LANGUAGE")

    label2 = Label(root, text="ENGLISH", font="segoe 30 bold", bg="white", width=18, bd=5, relief=GROOVE)
    label2.place(x=620, y=50)

    # First frame
    f = Frame(root, bg="Black", bd=5)
    f.place(x=10, y=118, width=440, height=210)

    text1 = Text(f, font="Roboto 20", bg="white", relief=GROOVE, wrap=WORD)
    text1.place(x=0, y=0, width=430, height=200)

    scrollbar1 = Scrollbar(f)
    scrollbar1.pack(side="right", fill='y')

    scrollbar1.configure(command=text1.yview)
    text1.configure(yscrollcommand=scrollbar1.set)

    # Second frame
    f1 = Frame(root, bg="Black", bd=5)
    f1.place(x=620, y=118, width=440, height=210)

    text2 = Text(f1, font="Roboto 20", bg="white", relief=GROOVE, wrap=WORD)
    text2.place(x=0, y=0, width=430, height=200)

    scrollbar2 = Scrollbar(f1)
    scrollbar2.pack(side="right", fill='y')

    scrollbar2.configure(command=text2.yview)
    text2.configure(yscrollcommand=scrollbar2.set)

    # Translate button
    translate = Button(root, text="Translate", font=("Roboto", 15), activebackground="white", cursor="hand2", bd=1,
                       width=10, height=2, bg="black", fg="white", command=translate_now)
    translate.place(x=476, y=150)

    label_change()
    
def textspeech():
    try:
        def speaknow():
            engine = pyttsx3.init()
            engine.setProperty('rate', 150)
            engine.setProperty('voice', combo_voice.get())
            engine.say(text1.get("1.0", END))
            engine.runAndWait()
            engine.stop()
      
        root = Tk()
        root.title("Text to speech") 
        root.geometry("400x200") 
        root.resizable(False, False) 

        obj = LabelFrame(root, text="Text to speech", font=20, bd=1)
        obj.pack(fill="both", expand="yes", padx=10, pady=10)

        lbl = Label(obj, text="Text", font=30) 
        lbl.pack(side=tk.LEFT, padx=5)
        text1 = Text(obj, font=30, width=25, bd=5) 
        text1.pack(side=tk.LEFT, padx=10)

        voice_var = StringVar()
        voices = pyttsx3.init().getProperty('voices')
        voice_options = [voice.name for voice in voices]
        combo_voice = ttk.Combobox(obj, textvariable=voice_var, values=voice_options)
        combo_voice.pack(side=tk.LEFT, padx=10)
        combo_voice.set(voice_options[0])
        
        def speak():
            speaknow()
            
        btn_speak = Button(obj, text="Speak", font=20, bg="black", fg="white", command=speak) 
        btn_speak.pack(side=tk.LEFT, padx=10) 
        
        def on_enter(event):
            speaknow()
        
        text1.bind("<Return>", on_enter)
        
    except Exception as e:
        print("Error:", e)

def filespeech():
    try:
        root = Tk()
        root.title("File to Speech") 
        root.geometry("500x300")
        root.resizable(False, False)
        root.configure(background="White")

        lbl = Label(root, text="Select a text file", bg="white")
        lbl.place(x=10, y=50)

        def file():
            file_path = filedialog.askopenfilename()
            if file_path:
                with open(file_path, "r") as file:
                    text = file.read()
                engine = pyttsx3.init()
                engine.setProperty('rate', 150)
                engine.setProperty('voice', combo_voice.get())
                engine.say(text)
                engine.runAndWait()
                engine.stop()
        
        btn_browse = Button(root, text="Browse", command=file)
        btn_browse.place(x=150, y=50)

        voice_var = StringVar()
        voices = pyttsx3.init().getProperty('voices')
        voice_options = [voice.name for voice in voices]
        combo_voice = ttk.Combobox(root, textvariable=voice_var, values=voice_options)
        combo_voice.place(x=250, y=50)
        combo_voice.set(voice_options[0])
        
    except Exception as e:
        print("Error:", e)

def speechtotext():
    try:
        recognizer = sr.Recognizer()

        def record_audio():
            with sr.Microphone() as source:
                print("Listening...")
                recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
                audio = recognizer.listen(source)
            return audio

        def recognize_speech():
            try:
                audio = record_audio()
                src_language = combo_language.get()
                src_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(src_language.lower())]
                text = recognizer.recognize_google(audio, language=src_language_code)
                print("Recognized text:", text)
                text_output.delete(1.0, END)
                text_output.insert(END, text)
                
                # Translate recognized text if translation language selected
                if combo_translate_language.get() != "None":
                    dest_language = combo_translate_language.get()
                    dest_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(dest_language.lower())]
                    translator = Translator()
                    translated_text = translator.translate(text, dest=dest_language_code)
                    text_translated.delete(1.0, END)
                    text_translated.insert(END, translated_text.text)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
                messagebox.showerror("Error", "Could not connect to the internet.")

        root = Tk()
        root.title("Speech to Text")
        root.geometry("700x600")
        root.resizable(False, False)
        root.configure(background="White")

        # Label for recognized speech
        label_recognized = Label(root, text="Recognized speech", font=("Arial", 14, "bold"))
        label_recognized.pack(pady=10)

        # Language selection for speech recognition
        language_var = StringVar()
        language_options = list(googletrans.LANGUAGES.values())
        combo_language = ttk.Combobox(root, textvariable=language_var, values=language_options)
        combo_language.pack(pady=5)
        combo_language.set("english")

        # Label for translated speech
        label_translated = Label(root, text="Translated speech", font=("Arial", 14, "bold"))
        label_translated.pack(pady=10)

        # Language selection for translation
        translate_language_var = StringVar()
        translate_language_options = ["None"] + language_options
        combo_translate_language = ttk.Combobox(root, textvariable=translate_language_var, values=translate_language_options)
        combo_translate_language.pack(pady=5)
        combo_translate_language.set("None")

        record_button = Button(root, text="Record", command=recognize_speech)
        record_button.pack(pady=5)

        text_output = Text(root, height=10, width=50)
        text_output.pack(pady=10)
        
        text_translated = Text(root, height=10, width=50)
        text_translated.pack(pady=10)

    except Exception as e:
        print("Error:", e)
        messagebox.showerror("Error", "An error occurred. Please try again.")

def speechtospeech():
    try:
        recognizer = sr.Recognizer()
        engine = pyttsx3.init()

        def record_audio():
            with sr.Microphone() as source:
                print("Listening...")
                recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
                audio = recognizer.listen(source)
            return audio

        def speaknow(text, language_code):
            voices = engine.getProperty('voices')
            for voice in voices:
                if language_code in voice.id:
                    engine.setProperty('voice', voice.id)
                    break
            engine.say(text)
            engine.runAndWait()
            engine.stop()

        def recognize_and_translate_speech():
            try:
                audio = record_audio()
                src_language = combo_language.get()
                src_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(src_language.lower())]
                text = recognizer.recognize_google(audio, language=src_language_code)
                print("Recognized text:", text)
                text_output.delete(1.0, END)
                text_output.insert(END, text)
                
                # Translate recognized text
                dest_language = combo_translate_language.get()
                dest_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(dest_language.lower())]
                translator = Translator()
                translated_text = translator.translate(text, dest=dest_language_code).text
                print("Translated text:", translated_text)
                text_translated.delete(1.0, END)
                text_translated.insert(END, translated_text)
                
                # Speak translated text
                speaknow(translated_text, dest_language_code)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
                messagebox.showerror("Error", "Could not connect to the internet.")

        root = Tk()
        root.title("Speech to Speech")
        root.geometry("700x500")
        root.resizable(False, False)
        root.configure(background="White")

        # Label for recognized speech
        label_recognized = Label(root, text="Speech to Speech Translation", font=("Arial", 14, "bold"))
        label_recognized.pack(pady=10)

        # Language selection for speech recognition
        language_var = StringVar()
        language_options = list(googletrans.LANGUAGES.values())
        combo_language = ttk.Combobox(root, textvariable=language_var, values=language_options)
        combo_language.pack(pady=5)
        combo_language.set("english")

        # Language selection for translation
        translate_language_var = StringVar()
        translate_language_options = language_options
        combo_translate_language = ttk.Combobox(root, textvariable=translate_language_var, values=translate_language_options)
        combo_translate_language.pack(pady=5)
        combo_translate_language.set("spanish")

        record_button = Button(root, text="Translate Speech", command=recognize_and_translate_speech)
        record_button.pack(pady=20)

        text_output = Text(root, height=10, width=50)
        text_output.pack(pady=10)
        
        text_translated = Text(root, height=10, width=50)
        text_translated.pack(pady=10)

    except Exception as e:
        print("Error:", e)
        messagebox.showerror("Error", "An error occurred. Please try again.")


button_1 = customtkinter.CTkButton(master=frame,
                                   text="Text-to-Text",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=texttrans,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_1.pack(padx=20, pady=20)

button_2 = customtkinter.CTkButton(master=frame,
                                   text="Text-to-Speech",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=textspeech,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E",
                                  )

button_2.pack(padx=20, pady=20)

button_3 = customtkinter.CTkButton(master=frame,
                                   text="Speech-to-Text",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=speechtotext,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_3.pack(padx=20, pady=20)

button_4 = customtkinter.CTkButton(master=frame,
                                   text="File-to-Speech",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=filespeech,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_4.pack(padx=20, pady=20)

button_5 = customtkinter.CTkButton(master=frame,
                                   text="Speech-to-Speech",
                                   corner_radius=8,
                                   width=200,
                                   height=30,
                                   command=speechtospeech,
                                   font=("Arial", 20),
                                   border_spacing=10,
                                   fg_color="#D2122E")
button_5.pack(padx=20, pady=20)

app.mainloop()
#Final
```

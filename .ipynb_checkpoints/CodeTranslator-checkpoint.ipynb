```python
from tkinter import *  # Importing tkinter for GUI elements
import tkinter as tk  # Importing tkinter as tk for convenience
from tkinter import ttk, messagebox, filedialog  # Importing specific tkinter modules
import googletrans  # Importing googletrans for language translation
from googletrans import Translator  # Importing Translator class from googletrans
import pyttsx3  # Importing pyttsx3 for text-to-speech functionality
import speech_recognition as sr  # Importing speech_recognition for speech-to-text functionality
import customtkinter  # Importing customtkinter for custom styled tkinter widgets
from PIL import Image # Import the Image class from the PIL (Pillow) library.The PIL library is used for opening, manipulating, and saving many different image file formats.
import pytesseract # pytesseract is a Python wrapper for Google's Tesseract-OCR Engine, which is used for Optical Character Recognition (OCR).

#Update this path to the location of the Tesseract executable on your system
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def imagetotext_speech():
    try:
        # Create the main window for the application
        root = tk.Tk()
        root.title("Image to Text Converter")  # Set the title of the window
        root.geometry("600x500")  # Set the size of the window
        root.resizable(False, False)  # Prevent the user from resizing the window
        root.configure(background="white")  # Set the background color of the window

        # Create a frame for buttons and voice selection
        frame_buttons = tk.Frame(root, bg="white")
        frame_buttons.pack(pady=10, fill=tk.X)

        # Create a frame for text output
        frame_text = tk.Frame(root, bg="white")
        frame_text.pack(pady=10, fill=tk.BOTH, expand=True)

        # Create a frame to contain both the buttons and text output for better organization
        frame_main = tk.Frame(root)
        frame_main.pack(fill=tk.BOTH, expand=True)

        # Add instruction label
        lbl = tk.Label(frame_buttons, text="Select an image and convert it to text", bg="white")
        lbl.pack(pady=10)

        # Text widget to display the extracted text
        text_output = tk.Text(frame_text, height=15, width=70, wrap=tk.WORD)
        text_output.pack(pady=10, fill=tk.BOTH, expand=True)
        
        # Scrollbar for the text widget
        scrollbar = tk.Scrollbar(frame_text, command=text_output.yview)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        text_output.config(yscrollcommand=scrollbar.set)

        # Function to handle image selection and text extraction
        def select_image():
            # Open a file dialog to select an image file
            file_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.png;*.jpg;*.jpeg;*.bmp")])
            if file_path:  # Check if a file was selected
                try:
                    # Load the image using PIL
                    image = Image.open(file_path)
                    # Use pytesseract to extract text from the image
                    extracted_text = pytesseract.image_to_string(image)
                    
                    if not extracted_text.strip():
                        messagebox.showinfo("Result", "No text found in the image.")
                        return
                    
                    # Display the extracted text in the text widget
                    text_output.delete(1.0, tk.END)
                    text_output.insert(tk.END, extracted_text)
                    
                except Exception as e:
                    print("Error:", e)
                    messagebox.showerror("Error", "An error occurred while processing the image.")
        
        # Function to convert the extracted text to speech
        def speak_text():
            extracted_text = text_output.get(1.0, tk.END).strip()
            if not extracted_text:
                messagebox.showinfo("Result", "No text available. Please select an image first.")
                return
            
            try:
                # Initialize the text-to-speech engine
                engine = pyttsx3.init()
                engine.setProperty('rate', 150)  # Set the speech rate
                engine.setProperty('voice', combo_voice.get())  # Set the voice based on user selection
                engine.say(extracted_text)  # Convert the text to speech
                engine.runAndWait()  # Wait for the speech to finish
                engine.stop()  # Stop the speech engine

            except Exception as e:
                print("Error:", e)
                messagebox.showerror("Error", "An error occurred while converting text to speech.")

        # Create buttons for selecting image and speaking text
        btn_select_image = tk.Button(frame_buttons, text="Select Image", command=select_image)
        btn_select_image.pack(side=tk.LEFT, padx=10)

        btn_speak_text = tk.Button(frame_buttons, text="Speak Text", command=speak_text)
        btn_speak_text.pack(side=tk.LEFT, padx=10)

        # Variable to store the selected voice
        voice_var = tk.StringVar()
        # Initialize the text-to-speech engine to get available voices
        voices = pyttsx3.init().getProperty('voices')
        # Extract the names of all available voices
        voice_options = [voice.name for voice in voices]
        # Create a combo box to allow the user to select a voice
        combo_voice = ttk.Combobox(frame_buttons, textvariable=voice_var, values=voice_options)
        combo_voice.pack(side=tk.RIGHT, padx=10)
        combo_voice.set(voice_options[0])  # Set the default selected voice to the first in the list

        root.mainloop()  # Run the main loop of the application

    except Exception as e:
        # Catch any exceptions that occur and print an error message
        print("Error:", e)  # Print the exception details

# Setting appearance and color theme for customtkinter
customtkinter.set_appearance_mode("dark")  # Set the appearance mode to dark
customtkinter.set_default_color_theme("green")  # Set the default color theme to green

# Initializing main application window
app = customtkinter.CTk()  # Create the main application window
app.title("Helping Hand")  # Set the title of the application window
app.geometry("800x600")  # Set the size of the application window

# Creating a frame to hold the main interface
frame = customtkinter.CTkFrame(master=app, corner_radius=20)  # Create a frame with rounded corners
Label(frame, text="Helping Hands- A Dynamic Translator", font="ariel 25 bold", fg="navy", bg="grey").pack(fill="both")
# Add a label to the frame with a title
frame.pack(pady=20, padx=20, fill="both", expand=True)  # Pack the frame with padding and make it expandable

# Function to handle text-to-text translation
def texttrans():
    # Create a new window for the text-to-text translation feature
    root = Tk()  # Initialize a new Tkinter window
    root.title("Google Translator")  # Set the title of the new window
    root.geometry("1080x400")  # Set the size of the new window
    root.resizable(False, False)  # Make the window non-resizable
    root.configure(background="White")  # Set the background color of the new window

    # Function to update the language labels periodically
    def label_change():
        c = combo1.get()  # Get the selected source language from the combo box
        c1 = combo2.get()  # Get the selected target language from the combo box
        label1.configure(text=c.upper())  # Update the source language label
        label2.configure(text=c1.upper())  # Update the target language label
        root.after(1000, label_change)  # Call this function again after 1 second

    # Function to translate the text
    def translate_now():
        text_ = text1.get(1.0, END)  # Get the input text from the text widget
        t1 = Translator()  # Initialize the Google Translator
        # Translate the text from source language to target language
        trans_text = t1.translate(text_, dest=combo2.get(), src=combo1.get())
        trans_text = trans_text.text  # Extract the translated text

        text2.delete(1.0, END)  # Clear the target text widget
        text2.insert(END, trans_text)  # Insert the translated text into the target text widget

    # Get available languages from Google Translator
    language = googletrans.LANGUAGES  # Dictionary of language codes and their names
    languageV = list(language.values())  # List of language names
    lang1 = language.keys()  # List of language codes

    # First combo box for source language selection
    combo1 = ttk.Combobox(root, values=languageV, font="Roboto 14", state="r")
    combo1.place(x=110, y=20)  # Place the combo box at specific coordinates
    combo1.set("ENGLISH")  # Set the default value of the combo box

    # Label for source language
    label1 = Label(root, text="ENGLISH", font="segoe 30 bold", bg="white", width=18, bd=5, relief=GROOVE)
    label1.place(x=10, y=50)  # Place the label at specific coordinates

    # Second combo box for target language selection
    combo2 = ttk.Combobox(root, values=languageV, font="Roboto 14", state="r")
    combo2.place(x=730, y=20)  # Place the combo box at specific coordinates
    combo2.set("SELECT LANGUAGE")  # Set the default value of the combo box

    # Label for target language
    label2 = Label(root, text="ENGLISH", font="segoe 30 bold", bg="white", width=18, bd=5, relief=GROOVE)
    label2.place(x=620, y=50)  # Place the label at specific coordinates

    # Frame for input text
    f = Frame(root, bg="Black", bd=5)
    f.place(x=10, y=118, width=440, height=210)  # Place the frame at specific coordinates

    text1 = Text(f, font="Roboto 20", bg="white", relief=GROOVE, wrap=WORD)
    text1.place(x=0, y=0, width=430, height=200)  # Place the text widget inside the frame

    # Scrollbar for the input text widget
    scrollbar1 = Scrollbar(f)
    scrollbar1.pack(side="right", fill='y')  # Pack the scrollbar to the right side

    scrollbar1.configure(command=text1.yview)  # Configure the scrollbar to work with the text widget
    text1.configure(yscrollcommand=scrollbar1.set)  # Set the scrollbar command for the text widget

    # Frame for translated text
    f1 = Frame(root, bg="Black", bd=5)
    f1.place(x=620, y=118, width=440, height=210)  # Place the frame at specific coordinates

    text2 = Text(f1, font="Roboto 20", bg="white", relief=GROOVE, wrap=WORD)
    text2.place(x=0, y=0, width=430, height=200)  # Place the text widget inside the frame

    # Scrollbar for the translated text widget
    scrollbar2 = Scrollbar(f1)
    scrollbar2.pack(side="right", fill='y')  # Pack the scrollbar to the right side

    scrollbar2.configure(command=text2.yview)  # Configure the scrollbar to work with the text widget
    text2.configure(yscrollcommand=scrollbar2.set)  # Set the scrollbar command for the text widget

    # Button to initiate translation
    translate = Button(root, text="Translate", font=("Roboto", 15), activebackground="white", cursor="hand2", bd=1,
                       width=10, height=2, bg="black", fg="white", command=translate_now)
    translate.place(x=476, y=150)  # Place the button at specific coordinates

    label_change()  # Call the function to update the language labels periodically

# Function to handle text-to-speech conversion
def textspeech():
    try:
        # Function to perform the text-to-speech conversion
        def speaknow():
            # Initialize the text-to-speech engine
            engine = pyttsx3.init()
            engine.setProperty('rate', 150)  # Set the speech rate (words per minute)
            engine.setProperty('voice', combo_voice.get())  # Set the voice based on user selection
            # Convert the text from the Text widget to speech
            engine.say(text1.get("1.0", END))  # Retrieve the entire content of the Text widget
            engine.runAndWait()  # Wait for the speech to finish
            engine.stop()  # Stop the speech engine

        # Create the main window for the application
        root = Tk()
        root.title("Text to Speech")  # Set the title of the window
        root.geometry("400x200")  # Set the size of the window
        root.resizable(False, False)  # Prevent the user from resizing the window

        # Create a LabelFrame to contain the widgets
        obj = LabelFrame(root, text="Text to Speech", font=20, bd=1)
        obj.pack(fill="both", expand="yes", padx=10, pady=10)  # Pack the LabelFrame with padding

        # Create a Label widget to prompt the user for text input
        lbl = Label(obj, text="Text", font=30) 
        lbl.pack(side=tk.LEFT, padx=5)  # Pack the label on the left side with padding

        # Create a Text widget for text input
        text1 = Text(obj, font=30, width=25, bd=5)
        text1.pack(side=tk.LEFT, padx=10)  # Pack the Text widget on the left side with padding

        # Create a StringVar to store the selected voice
        voice_var = StringVar()
        # Initialize the text-to-speech engine to get available voices
        voices = pyttsx3.init().getProperty('voices')
        # Extract the names of all available voices
        voice_options = [voice.name for voice in voices]
        # Create a combo box to allow the user to select a voice
        combo_voice = ttk.Combobox(obj, textvariable=voice_var, values=voice_options)
        combo_voice.pack(side=tk.LEFT, padx=10)  # Pack the combo box on the left side with padding
        combo_voice.set(voice_options[0])  # Set the default selected voice to the first in the list
        
        # Function to call speaknow() when the button is pressed
        def speak():
            speaknow()
            
        # Create a Button to start the text-to-speech conversion
        btn_speak = Button(obj, text="Speak", font=20, bg="black", fg="white", command=speak)
        btn_speak.pack(side=tk.LEFT, padx=10)  # Pack the button on the left side with padding
        
        # Function to call speaknow() when the Enter key is pressed in the Text widget
        def on_enter(event):
            speaknow()
        
        # Bind the Enter key press event to the on_enter function
        text1.bind("<Return>", on_enter)
        
    except Exception as e:
        # Catch any exceptions that occur and print an error message
        print("Error:", e)

# Function to handle file-to-speech conversion
def filespeech():
    try:
        # Create the main window for the application
        root = Tk()
        root.title("File to Speech")  # Set the title of the window
        root.geometry("500x300")  # Set the size of the window
        root.resizable(False, False)  # Prevent the user from resizing the window
        root.configure(background="White")  # Set the background color of the window

        # Label to prompt the user to select a text file
        lbl = Label(root, text="Select a text file", bg="white")
        lbl.place(x=10, y=50)  # Place the label at the specified coordinates

        # Function to handle file selection and speech conversion
        def file():
            # Open a file dialog to select a text file
            file_path = filedialog.askopenfilename()
            if file_path:  # Check if a file was selected
                with open(file_path, "r") as file:  # Open the selected file for reading
                    text = file.read()  # Read the entire content of the file
                # Initialize the text-to-speech engine
                engine = pyttsx3.init()
                engine.setProperty('rate', 150)  # Set the speech rate (words per minute)
                engine.setProperty('voice', combo_voice.get())  # Set the voice based on user selection
                engine.say(text)  # Convert the text to speech
                engine.runAndWait()  # Wait for the speech to finish
                engine.stop()  # Stop the speech engine

        # Button to open file dialog and start the file processing
        btn_browse = Button(root, text="Browse", command=file)
        btn_browse.place(x=150, y=50)  # Place the button at the specified coordinates

        # Variable to store the selected voice
        voice_var = StringVar()
        # Initialize the text-to-speech engine to get available voices
        voices = pyttsx3.init().getProperty('voices')
        # Extract the names of all available voices
        voice_options = [voice.name for voice in voices]
        # Create a combo box to allow the user to select a voice
        combo_voice = ttk.Combobox(root, textvariable=voice_var, values=voice_options)
        combo_voice.place(x=250, y=50)  # Place the combo box at the specified coordinates
        combo_voice.set(voice_options[0])  # Set the default selected voice to the first in the list

    except Exception as e:
        # Catch any exceptions that occur and print an error message
        print("Error:", e)  # Print the exception details
        
# Function to handle speech-to-text conversion
def speechtotext():
    try:
        # Initialize the recognizer for speech recognition
        recognizer = sr.Recognizer()

        # Function to record audio from the microphone
        def record_audio():
            with sr.Microphone() as source:
                print("Listening...")  # Inform the user that the microphone is listening
                recognizer.adjust_for_ambient_noise(source)  # Adjust the recognizer to account for ambient noise
                audio = recognizer.listen(source)  # Capture audio from the microphone
            return audio  # Return the captured audio

        # Function to recognize and optionally translate speech
        def recognize_speech():
            try:
                # Record audio from the microphone
                audio = record_audio()
                
                # Get the selected source language from the combo box
                src_language = combo_language.get()
                # Map the source language name to its language code
                src_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(src_language.lower())]
                
                # Recognize the text from the recorded audio using Google Speech Recognition
                text = recognizer.recognize_google(audio, language=src_language_code)
                print("Recognized text:", text)  # Print the recognized text
                
                # Update the text output widget with the recognized text
                text_output.delete(1.0, END)
                text_output.insert(END, text)
                
                # Check if translation is enabled (i.e., if a destination language is selected)
                if combo_translate.get() != "None":
                    # Get the selected target language from the combo box
                    dest_language = combo_translate_language.get()
                    # Map the target language name to its language code
                    dest_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(dest_language.lower())]
                    
                    # Initialize the translator and translate the recognized text
                    translator = Translator()
                    translated_text = translator.translate(text, dest=dest_language_code)
                    
                    # Update the text widget with the translated text
                    text_translated.delete(1.0, END)
                    text_translated.insert(END, translated_text.text)
            except sr.UnknownValueError:
                # Handle case where audio is not recognized
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                # Handle errors in connecting to the Google Speech Recognition service
                print(f"Could not request results from Google Speech Recognition service; {e}")
                messagebox.showerror("Error", "Could not connect to the internet.")  # Show an error message to the user

        # Create the main window for the application
        root = Tk()
        root.title("Speech to Text")  # Set the title of the window
        root.geometry("700x600")  # Set the size of the window
        root.resizable(False, False)  # Prevent the user from resizing the window
        root.configure(background="White")  # Set the background color of the window

        # Label for displaying the recognized speech
        label_recognized = Label(root, text="Recognized speech", font=("Arial", 14, "bold"))
        label_recognized.pack(pady=10)  # Add the label to the window with vertical padding

        # Create a variable to store the selected language for speech recognition
        language_var = StringVar()
        # List of language options for speech recognition
        language_options = list(googletrans.LANGUAGES.values())
        # Create a combo box for selecting the source language
        combo_language = ttk.Combobox(root, textvariable=language_var, values=language_options)
        combo_language.pack(pady=5)  # Add the combo box to the window with vertical padding
        combo_language.set("english")  # Set the default selected language to English

        # Label for displaying the translated speech
        label_translated = Label(root, text="Translated speech", font=("Arial", 14, "bold"))
        label_translated.pack(pady=10)  # Add the label to the window with vertical padding

        # Create a variable to store the selected language for translation
        translate_language_var = StringVar()
        # Add "None" option to the list of languages for translation (no translation)
        translate_language_options = ["None"] + language_options
        # Create a combo box for selecting the target language for translation
        combo_translate_language = ttk.Combobox(root, textvariable=translate_language_var, values=translate_language_options)
        combo_translate_language.pack(pady=5)  # Add the combo box to the window with vertical padding
        combo_translate_language.set("None")  # Set the default selection to "None" (no translation)

        # Button to start the process of recording and recognizing speech
        record_button = Button(root, text="Record", command=recognize_speech)
        record_button.pack(pady=5)  # Add the button to the window with vertical padding

        # Text widget to display recognized speech
        text_output = Text(root, height=10, width=50)
        text_output.pack(pady=10)  # Add the text widget to the window with vertical padding

        # Text widget to display translated text (if applicable)
        text_translated = Text(root, height=10, width=50)
        text_translated.pack(pady=10)  # Add the text widget to the window with vertical padding

    except Exception as e:
        # Catch any exceptions that occur and display an error message
        print("Error:", e)  # Print the exception details
        messagebox.showerror("Error", "An error occurred. Please try again.")  # Show an error message to the user

# Function to handle speech-to-speech translation
def speechtospeech():
    try:
        # Initialize the recognizer for speech recognition
        recognizer = sr.Recognizer()
        
        # Initialize the text-to-speech engine
        engine = pyttsx3.init()

        # Function to record audio from the microphone
        def record_audio():
            with sr.Microphone() as source:
                print("Listening...")  # Inform the user that the microphone is listening
                recognizer.adjust_for_ambient_noise(source)  # Adjust the recognizer to account for ambient noise
                audio = recognizer.listen(source)  # Capture audio from the microphone
            return audio  # Return the captured audio

        # Function to speak out the given text using the specified language
        def speaknow(text, language_code):
            # Get a list of available voices
            voices = engine.getProperty('voices')
            for voice in voices:
                # Set the voice if it matches the desired language code
                if language_code in voice.id:
                    engine.setProperty('voice', voice.id)
                    break
            engine.say(text)  # Pass the text to the speech engine
            engine.runAndWait()  # Wait until the speech is done
            engine.stop()  # Stop the speech engine

        # Function to recognize, translate, and speak the recognized speech
        def recognize_and_translate_speech():
            try:
                # Record audio from the microphone
                audio = record_audio()
                
                # Get the source language from the combo box
                src_language = combo_language.get()
                # Map the source language name to its language code
                src_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(src_language.lower())]
                
                # Recognize the text from the recorded audio using Google Speech Recognition
                text = recognizer.recognize_google(audio, language=src_language_code)
                print("Recognized text:", text)  # Print the recognized text
                # Update the text output widget with the recognized text
                text_output.delete(1.0, END)
                text_output.insert(END, text)

                # Get the target language from the combo box
                dest_language = combo_translate_language.get()
                # Map the target language name to its language code
                dest_language_code = list(googletrans.LANGUAGES.keys())[list(googletrans.LANGUAGES.values()).index(dest_language.lower())]
                
                # Initialize the translator and translate the recognized text
                translator = Translator()
                translated_text = translator.translate(text, dest=dest_language_code).text
                print("Translated text:", translated_text)  # Print the translated text
                # Update the text widget with the translated text
                text_translated.delete(1.0, END)
                text_translated.insert(END, translated_text)

                # Speak the translated text in the selected language
                speaknow(translated_text, dest_language_code)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")  # Handle case where audio is not recognized
            except sr.RequestError as e:
                print(f"Could not request results from Google Speech Recognition service; {e}")  # Handle connection errors
                messagebox.showerror("Error", "Could not connect to the internet.")  # Show an error message to the user

        # Create the main window for the application
        root = Tk()
        root.title("Speech to Speech")  # Set the title of the window
        root.geometry("700x500")  # Set the size of the window
        root.resizable(False, False)  # Prevent the user from resizing the window
        root.configure(background="White")  # Set the background color of the window

        # Label to display the title or description of the application
        label_recognized = Label(root, text="Speech to Speech Translation", font=("Arial", 14, "bold"))
        label_recognized.pack(pady=10)  # Add the label to the window with vertical padding

        # Create a variable to store the selected language for speech recognition
        language_var = StringVar()
        # List of language options for speech recognition
        language_options = list(googletrans.LANGUAGES.values())
        # Create a combo box for selecting the source language
        combo_language = ttk.Combobox(root, textvariable=language_var, values=language_options)
        combo_language.pack(pady=5)  # Add the combo box to the window with vertical padding
        combo_language.set("english")  # Set the default selected language to English

        # Create a variable to store the selected language for translation
        translate_language_var = StringVar()
        # List of language options for translation (same as source language options)
        translate_language_options = language_options
        # Create a combo box for selecting the target language
        combo_translate_language = ttk.Combobox(root, textvariable=translate_language_var, values=translate_language_options)
        combo_translate_language.pack(pady=5)  # Add the combo box to the window with vertical padding
        combo_translate_language.set("spanish")  # Set the default selected language to Spanish

        # Button to start the process of recording, recognizing, and translating speech
        record_button = Button(root, text="Translate Speech", command=recognize_and_translate_speech)
        record_button.pack(pady=20)  # Add the button to the window with vertical padding

        # Text widget to display the recognized speech
        text_output = Text(root, height=10, width=50)
        text_output.pack(pady=10)  # Add the text widget to the window with vertical padding

        # Text widget to display the translated text
        text_translated = Text(root, height=10, width=50)
        text_translated.pack(pady=10)  # Add the text widget to the window with vertical padding

    except Exception as e:
        # Catch any exceptions that occur and display an error message
        print("Error:", e)  # Print the exception details
        messagebox.showerror("Error", "An error occurred. Please try again.")  # Show an error message to the user

# Creating a button for the "Text-to-Text" functionality on the main interface
# This button allows the user to convert text to text using the specified function `texttrans`
button_1 = customtkinter.CTkButton(
    master=frame,                # The parent container where the button will be placed
    text="Text-to-Text",         # The text label displayed on the button
    corner_radius=8,             # The radius of the button corners (for rounded corners)
    width=200,                   # The width of the button in pixels
    height=30,                   # The height of the button in pixels
    command=texttrans,           # The function to be called when the button is clicked
    font=("Arial", 20),          # The font style and size for the button text
    border_spacing=10,           # The spacing between the button border and the text
    fg_color="#D2122E"           # The foreground color of the button (text color)
)
button_1.pack(padx=20, pady=20)   # Adding the button to the layout with padding around it

# Creating a button for the "Text-to-Speech" functionality on the main interface
# This button allows the user to convert text to speech using the specified function `textspeech`
button_2 = customtkinter.CTkButton(
    master=frame,                # The parent container where the button will be placed
    text="Text-to-Speech",       # The text label displayed on the button
    corner_radius=8,             # The radius of the button corners (for rounded corners)
    width=200,                   # The width of the button in pixels
    height=30,                   # The height of the button in pixels
    command=textspeech,          # The function to be called when the button is clicked
    font=("Arial", 20),          # The font style and size for the button text
    border_spacing=10,           # The spacing between the button border and the text
    fg_color="#D2122E"           # The foreground color of the button (text color)
)
button_2.pack(padx=20, pady=20)   # Adding the button to the layout with padding around it

# Creating a button for the "Speech-to-Text" functionality on the main interface
# This button allows the user to convert speech to text using the specified function `speechtotext`
button_3 = customtkinter.CTkButton(
    master=frame,                # The parent container where the button will be placed
    text="Speech-to-Text",       # The text label displayed on the button
    corner_radius=8,             # The radius of the button corners (for rounded corners)
    width=200,                   # The width of the button in pixels
    height=30,                   # The height of the button in pixels
    command=speechtotext,        # The function to be called when the button is clicked
    font=("Arial", 20),          # The font style and size for the button text
    border_spacing=10,           # The spacing between the button border and the text
    fg_color="#D2122E"           # The foreground color of the button (text color)
)
button_3.pack(padx=20, pady=20)   # Adding the button to the layout with padding around it

# Creating a button for the "File-to-Speech" functionality on the main interface
# This button allows the user to convert a file to speech using the specified function `filespeech`
button_4 = customtkinter.CTkButton(
    master=frame,                # The parent container where the button will be placed
    text="File-to-Speech",       # The text label displayed on the button
    corner_radius=8,             # The radius of the button corners (for rounded corners)
    width=200,                   # The width of the button in pixels
    height=30,                   # The height of the button in pixels
    command=filespeech,          # The function to be called when the button is clicked
    font=("Arial", 20),          # The font style and size for the button text
    border_spacing=10,           # The spacing between the button border and the text
    fg_color="#D2122E"           # The foreground color of the button (text color)
)
button_4.pack(padx=20, pady=20)   # Adding the button to the layout with padding around it

# Creating a button for the "Speech-to-Speech" functionality on the main interface
# This button allows the user to convert speech to speech using the specified function `speechtospeech`
button_5 = customtkinter.CTkButton(
    master=frame,                # The parent container where the button will be placed
    text="Speech-to-Speech",     # The text label displayed on the button
    corner_radius=8,             # The radius of the button corners (for rounded corners)
    width=200,                   # The width of the button in pixels
    height=30,                   # The height of the button in pixels
    command=speechtospeech,      # The function to be called when the button is clicked
    font=("Arial", 20),          # The font style and size for the button text
    border_spacing=10,           # The spacing between the button border and the text
    fg_color="#D2122E"           # The foreground color of the button (text color)
)
button_5.pack(padx=20, pady=20)   # Adding the button to the layout with padding around it

# Creating a button for the "Image-to-Text&Speech" functionality on the main interface
# This button allows the user to convert Image to Text & Speech using the specified function `imagetotext`
button_6 = customtkinter.CTkButton(
    master=frame,                       # The parent container where the button will be placed
    text="Image-to-Text&Speech",        # The text label displayed on the button
    corner_radius=8,                    # The radius of the button corners (for rounded corners)
    width=200,                          # The width of the button in pixels
    height=30,                          # The height of the button in pixels
    command=imagetotext_speech,         # The function to be called when the button is clicked
    font=("Arial", 20),                 # The font style and size for the button text
    border_spacing=10,                  # The spacing between the button border and the text
    fg_color="#D2122E"                  # The foreground color of the button (text color)
)
button_6.pack(padx=20, pady=20)   # Adding the button to the layout with padding around it

# Starting the main application loop to make the GUI responsive
# This line keeps the application running and listens for user interactions
app.mainloop()
```
